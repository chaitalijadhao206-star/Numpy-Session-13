```python
import numpy as np
```

```python
# Get number of random numbers
n = int(input("Enter number of elements: "))

# Generate random numbers
arr = np.random.randint(10, 101, n)

# Print array
print("\nArray:")
print(arr)

# Print statistics
print("\nMean:", np.mean(arr))
print("\nMedian:", np.median(arr))
print("\nStandard Deviation:", np.std(arr))
print("\nMinimum:", np.min(arr))
print("\nMaximum:", np.max(arr))

# Reshape into 2D array if possible
if n % 2 == 0:
    arr2 = arr.reshape(2, n // 2)

    print("2D Array:")
    print(arr2)

    print("Row-wise Sum:")
    print(np.sum(arr2, axis=1))
else:
    print("Cannot reshape into 2D array")
```
