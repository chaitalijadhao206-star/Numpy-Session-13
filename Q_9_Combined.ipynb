```python
import numpy as np
```

```python
# Create 1D array of 20 random integers
arr = np.random.randint(1, 51, 20)

# Reshape into 4x5 matrix
mat = arr.reshape(4, 5)

# Print matrix
print("4x5 Matrix:")
print(mat)

print()

# Print sum
print("Sum:", np.sum(mat))

# Print mean
print("Mean:", np.mean(mat))

# Print standard deviation
print("Standard Deviation:", np.std(mat))

print()

# Print maximum value of each row
print("Maximum Value in Each Row:")
print(np.max(mat, axis=1))
```
