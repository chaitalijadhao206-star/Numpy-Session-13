```python
import numpy as np
```

```python
# Create 10 equally spaced numbers from 0 to 5
arr=np.linspace(0, 5, 10)

# Print first array
print("10 Equally Spaced Numbers:")
print(arr)

# Print length of first array
print("Length:", len(arr))

# Create 15 equally spaced numbers from -10 to 10
arr=np.linspace(-10, 10, 15)

# Print second array
print("15 Equally Spaced Numbers:")
print(arr)

# Print length of second array
print("Length:", len(arr))
```
