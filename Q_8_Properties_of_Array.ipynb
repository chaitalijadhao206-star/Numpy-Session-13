```python
import numpy as np
```

```python
# Create 4x4 random matrix
arr = np.random.randint(1, 101, (4, 4))

print("4x4 Matrix:")
print(arr)

# Print shape
print("\nShape:", arr.shape)

# Print dimension
print("\nDimension:", arr.ndim)

# Print total elements
print("\nSize:", arr.size)

# Print data type
print("\nData Type:", arr.dtype)

# Print minimum value
print("\nMinimum Value:", arr.min())

# Print maximum value
print("\nMaximum Value:", arr.max())
```
