```python
import numpy as np
```

```python
# Create 1D array of zeros
arr1 = np.zeros(8)

# Print 1D array
print("1D Array of Zeros:")
print(arr1)

# Create 2D array of ones
arr2 = np.ones((4, 4))

# Print 1D array
print("\n2D Array of ones:")
print(arr2)

# Create 3x3 matrix of zeros
arr3 = np.zeros((3, 3))

# Print 3x3 matrix
print("\n3x3 Matrix of Zeros:")
print(arr3)



```
