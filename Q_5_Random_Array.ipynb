```python
import numpy as np
```

```python
# Create 1D array of 10 random numbers
arr=np.random.rand(10)

# Print first array
print("Random Numbers (0 to 1):")
print(arr)

# Create 3x3 matrix of random numbers
arr=np.random.randn(3, 3)

# Print second array
print("\n3x3 Random Matrix:")
print(arr)

# Create 4x5 matrix of random integers
arr=np.random.randint(10, 51, (4, 5))

# Print third array
print("\n4x5 Random Integer Matrix:")
print(arr)
```
