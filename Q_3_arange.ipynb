```python
import numpy as np
```

```python
# Create array from 0 to 20
arr=np.arange(0, 21)

# Print first array
print("Numbers from 0 to 20:")
print(arr)

# Create even numbers from 10 to 50
arr=np.arange(10, 51, 2)

# Print second array
print("\nEven Numbers from 10 to 50:")
print(arr)


# Create numbers from 5 to 100 with step 5
arr=np.arange(5, 101, 5)

# Print third array
print("\nNumbers from 5 to 100:")
print(arr)



```
